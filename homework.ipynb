### Problem 1

What is the marginal distribution of green ratings?
Set `p1` to the marginal distribution of green ratings represented as a dictionary of probabilities.
Use the integer ratings as the keys in `p1`.

```python
# YOUR CHANGES HERE
# reminder: marginal distribution of just 1 column is taking the count for one value / total count
# start by getting the values and their counts for green_rating
green_counts = mango_data["green_rating"].value_counts()
# print(green_counts)
# values are 1,2,3,4,5

green_len = len(mango_data["green_rating"])
# print(green_len)
# total count is 8

# make it a dictionary using a comprehension
p1 = {rating: count /green_len 
      for rating, count in green_counts.items()}
```
